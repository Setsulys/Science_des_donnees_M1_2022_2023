Choix régions

```python
fig, ax = plt.subplots()

ax.set(title='Choix régions', xlabel='longitude', ylabel='latitude')

z3 = ax.scatter(data = df_acc.loc[df_acc['lat'] > 0].loc[df_acc['long'] > -10], 
               x='long', y='lat', 
               alpha=0.2,
               s=1,)

z2 = ax.scatter(data = df_acc.loc[(df_acc['lat'] > 44) & (df_acc['lat'] < 45.5)].loc[(df_acc['long'] > 1) & (df_acc['long'] < 3.5)], 
               x='long', y='lat', 
               alpha=0.2,
               s=1,
               c='yellow')

z = ax.scatter(data = df_acc.loc[(df_acc['lat'] > 48) & (df_acc['lat'] < 49.5)].loc[(df_acc['long'] > 1) & (df_acc['long'] < 3.5)], 
               x='long', y='lat', 
               alpha=0.2,
               s=1,
               c='red')

plt.show()
```

Répartition genre Paris

```python
data = df_acc.loc[(df_acc['lat'] > 48) & (df_acc['lat'] < 49.5)].loc[(df_acc['long'] > 1) & (df_acc['long'] < 3.5)]
print(data.shape)
data = data['sexe'].value_counts(normalize=True) * 100
data.plot.pie(title = 'Répartition genre Paris',
       labels=['Hommes', 'Femmes', 'Indéterminé'],
       autopct='%1.1f%%')
```

Répartition genre centre sud France

```python
data = df_acc.loc[(df_acc['lat'] > 44) & (df_acc['lat'] < 45.5)].loc[(df_acc['long'] > 1) & (df_acc['long'] < 3.5)]
print(data.shape)
data = data['sexe'].value_counts(normalize=True) * 100
data.plot.pie(title = 'Répartition genre centre Sud France',
       labels=['Hommes', 'Femmes', 'Indéterminé'],
       autopct='%1.1f%%')
```
