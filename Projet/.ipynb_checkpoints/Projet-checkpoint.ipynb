```python
#df_acc.loc[df_acc["lum"] == 1].shape[0] # 165781 Jour
#df_acc.loc[df_acc["lum"] == 2].shape[0] # 15302 Aube/crépuscule


df_acc.loc[df_acc["lum"] > 2]["lum"].value_counts() #53920 aux total Nuit
labels=['Plein jour',
'Crépuscule ou aube',
'Nuit sans éclairage public',
'Nuit avec éclairage public non allumé',
'Nuit avec éclairage public allumé']

fig, axes = plt.subplots(ncols=2, gridspec_kw={'width_ratios': [2, 6]}, figsize=(15,5))
ax = axes[0]
ax.pie(df_acc["lum"].value_counts(normalize=True) * 100, labels= labels,
       autopct='%1.1f%%')
ax.title.set_text("Condition d'éclairage")
labels = labels[1:]


ax = axes[1]
ax.pie(df_acc.loc[df_acc["lum"]!=1]["lum"].value_counts(normalize=True) * 100,
       labels = labels,
       autopct='%1.1f%%')
ax.title.set_text("Condition d'éclairage la nuit")
plt.show()
```

La majorité des accidents se produisant en plein jour s'explique par le simple fait que nous, êtres humains, vivons dans une société active le jour, donc avec plus beaucoup plus de trafic. De façon étonnante, il y a 1.5x plus d'accidents la nuit lorsque l'éclairage est allumé que si elle ne l'était pas.

```python
def test_model(model, X, y, n_repet):
    pred = np.arange(n_repet)
    for i in range(n_repet):
        train_x,test_x,train_y,test_y = train_test_split(X,y,train_size=0.75)
        reg=model.fit(train_x,train_y)
        pred[i] = mean_squared_error(test_y,reg.predict(test_x))
    return "Perte moyenne :"+ str (pred.mean()) + " Ecart-type: " + str(pred.std())
```
